```python
label_encoder = LabelEncoder()
```

```python
# Fit and transform the 'species' column
dataset['species_encoded'] = label_encoder.fit_transform(dataset['Species'])

# Display the result
print(dataset[['Species', 'species_encoded']])
print(f"\nUnique classes and their integer mapping: {list(label_encoder.classes_)}")
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df['species'] = iris.target_names[iris.target]
```

```python
# Apply one-hot encoding using pandas get_dummies()
df_encoded = pd.get_dummies(df, columns=['species'], drop_first=False).inplace=True

# Display the results
print(df_encoded)
```
